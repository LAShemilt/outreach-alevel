```python
# Decision Trees
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Categorizing Fruit Data set

fruit = {'name': ['lemon', 'pear',  'raspberry', 'apple', 'banana'],
         'is_round': [1,0,1,1,0],
         'is_small': [0,0,1,0,0],
         'is_sweet': [0,1,1,1,1],
         'is_red':   [0,1,1,1,1],
         'is_green': [0,1,0,1,0]}
fruit_df = pd.DataFrame(fruit)

print(fruit_df.head())
```

```python
# Training the Tree

tree = DecisionTreeClassifier(max_leaf_nodes=5, random_state=0)
fit = tree.fit(fruit_df[['is_round','is_small','is_sweet', 'is_red', 'is_green']],fruit_df.name)

plot_tree(fit, class_names=fruit_df.name,
          feature_names= fruit_df.columns.to_list()[1:],
          filled=True)
plt.show()
```

```python
# Lets see if we can categorize some more fruit

new_fruit = {'name': ['Red Williams Pear', 'Granny Smith Apple', 'blackcurrent',
                      'pineapple'],
         'is_round': [0,1,1,0],
         'is_small': [0,0,1,1],
         'is_sweet': [1,1,1,1],
         'is_red':   [1,0,0,0],
         'is_green': [1,1,0,0]}
new_fruit_df = pd.DataFrame(new_fruit)

print(new_fruit_df.head())
```

```python
# What Categories do we predict?

output= tree.predict(new_fruit_df[['is_round','is_small','is_sweet', 'is_red', 'is_green']])
print(output)
```
